```python
sampleData = pd.read_csv('bookstore_inventory.csv')
sampleData
```

```python
plt.figure(figsize = (10, 6))
plt.barh(sampleData["qty"], sampleData["unit_price"])
plt.title("Bookstore Inventory")
plt.xlabel("Quantity")
plt.ylabel("Unit Price")
plt.show()
plt.grid()
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = sampleData["qty"], y = sampleData["unit_price"])
plt.title("Bookstore Inventory")
plt.xlabel("Quantity")
plt.ylabel("Unit Price")
plt.show()
plt.grid()
```
